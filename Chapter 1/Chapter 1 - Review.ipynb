## What is Machine Learning?
Is the science and art of programming a computer so they can learn from data.
A computer is said to learn from experience E with respect to some task T and some performance measure P, if its performance on T, as measured by P,  is improved by E.
So, we expect the task T (such as spam filter) to yield better performance P when it is exposed to some training data as experience E.

## Why use Machine Learning?
It can help on problems that require a lot of fine-tuning or long lists of rules.
For complex problems where traditional approach couldn’t help, ML can be implemented and potentially find a better alternative.
ML can adapt to new data (fluctuating environments)
It can give us insight about complex problems and large amounts of data.

## Types of Machine Learning Systems

### Supervised Learning

The training set includes the desired solutions, called labels.
Typical examples: classification (spam/ham) and regression (car price prediction)
algorithms: k-Nearest Neighbors, Linear Regression, Logistic Regression, Support Vector Machines (SVM), Decision Trees and Random Forests, Neural Networks (some NN can be unsupervised and semiunsupervided).

### Unsupervised Learning

The training data is unlabelled.
Clustering: K-means, DBSCAN, Hierarchical Cluster Analysis (HCA)
Anomaly detection and novelty: One-class SVM, Isolation Forest
Visualization and dimensionality reduction:Principal Component Analysis (PCA), Kernel PCA, Locally Linear Embedding (LLE), t-Distributed Stochastic Neighbor Embedding (t-SNE)
Association rule learning: Apriori, Eclat

### Semisupervised Learning

When the algorithm needs to learn with labelled and unlabelled instances in the data.
Deep belief networks (DBNs) = restricted Boltzmann machines (RBMs) stacked up + supervised learning for fine tuning.

### Reinforcement Learning

The learning system (agent) observes the environment, selects an action and gets rewarded or penalized.<br>
It then learns the best strategy (policy) by itself, to get the most reward over time.


### Batch and Online Learning
Whether or not a system can learn incrementally from a stream of incoming data

#### Batch Learning

Incapable or learning incrementally
Must be trained using all available data, taking much time and computer power, before being deployed to production.
Once on production, it only applies what it learnt. No more learning.
Offline learning
If new data appears, the whole process of training, evaluating and launching will have to be made from scratch, using new data + old data. Automation pipelines can help, but it is still might demand a lot of time and resources, depending on the application.

#### Online Learning

Opposite of batch learning.
Good for systems that receive data on continuous flow (stock prices) and need to adapt fast.
It can be used to train systems on huge datasets that cannot fit in one machine’s main memory (out-of-core learning). The algorithm loads part of the data, runs a training step on that data, and repeats it until it finishes the data.
Note that out-of-core learning are usually done offline. Think of it as incremental learning.

### Instance-Based x Model-Based Learning

#### Instance-Based Learning: 
“Learn by heart”, then generalizes to new cases by using a similarity measures to compare them to learned examples (or subset of them)

#### Model-Based Learning:
Generalize by building a model out of the examples and then use this model to make predictions.

## Testing and Validating

Split data in training and test sets, so you can validate your model against the test set and get the generalization error (or out-of-sample error). This value suggests how the model will perform on sets never seen before.
